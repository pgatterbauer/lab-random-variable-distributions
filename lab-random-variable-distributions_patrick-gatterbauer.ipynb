```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
numericals = df.select_dtypes(np.number)
```

```python
#Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
```

```python
for column in numericals:
    numericals[column].hist()
    plt.title(column)
    plt.show()
```

```python
for col in numericals:
    sns.distplot(numericals[col])
    plt.show()
```

```python
#Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
```

```python
#Normalizing data with standardscaler
```

```python
transformer = StandardScaler().fit(numericals)
num_normalized = transformer.transform(numericals)
x = pd.DataFrame(num_normalized)
x.columns = numericals.columns
x.head()
```

```python
for column in x:
    x[column].hist()
    plt.title(column)
    plt.show()
```

```python
#Normalizing data with MinMaxScaler
```

```python
scaler = MinMaxScaler()
# transform data
scaled = scaler.fit_transform(numericals)
print(scaled)
```

```python
num_mm = pd.DataFrame(scaled)
```

```python
num_mm
```

```python
for column in num_mm:
    num_mm[column].hist()
    plt.title(column)
    plt.show()
```

```python
#function for the log transform to replace infinite number
def log_transfom_clean_(x):
    if np.isfinite(x) and x!=0: # If the value is finite and != 0...
        return np.log(x)
    else:
        return np.NAN # We are returning NaNs so that we can replace them later
```

```python
for col in numericals:
    pd.Series(map(log_transfom_clean_, numericals[col])).hist()
    plt.title(col)
    plt.show()
```
